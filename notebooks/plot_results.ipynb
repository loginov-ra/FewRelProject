```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_scores = np.array(pd.read_csv('result_trn.csv', sep=';', header=None)).ravel()
val_scores = np.array(pd.read_csv('result_val.csv', sep=';', header=None)).ravel()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(val_scores, label='validation accuracy', c='purple')
plt.plot(train_scores, label='training accuracy')
plt.grid(ls=':')
plt.legend()
plt.title('Protonet model learning curve');
```
